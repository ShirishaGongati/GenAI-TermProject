```python
import requests
import json

def ollama_chat(prompt, model="llama3"):
    url = "http://localhost:11434/api/chat"
    
    payload = {
        "model": model,
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "stream": True   # IMPORTANT for streaming output
    }
    
    response = requests.post(url, json=payload, stream=True)
    
    final_output = ""
    
    for line in response.iter_lines():
        if line:
            try:
                data = json.loads(line.decode("utf-8"))
                if "message" in data and "content" in data["message"]:
                    final_output += data["message"]["content"]
            except Exception as e:
                print("Error while reading line:", line, e)
                
    return final_output
```

```python
output = ollama_chat("Say Hello World", model="llama3")
print(output)

```

```python
import ollama
import ipywidgets as widgets
from IPython.display import display, clear_output

# -----------------------
# FUNCTION TO CALL OLLAMA
# -----------------------
def generate_test_cases(user_story):
    prompt = f"""
You are a Senior QA Automation Engineer.
Generate detailed test cases for the following user story:

USER STORY:
{user_story}

Include the following in each test case:
1. Test Case ID
2. Test Title
3. Preconditions
4. Test Steps
5. Expected Result

Also include:
- Positive test cases
- Negative test cases
- Edge cases
- Validation scenarios
- Boundary conditions

Format clearly and professionally.
"""

    response = ollama.chat(
        model="llama3.1",
        messages=[{"role": "user", "content": prompt}]
    )

    return response['message']['content']


# -----------------------
# GUI WIDGETS
# -----------------------
input_box = widgets.Textarea(
    placeholder='Enter user story here...',
    description='Input:',
    layout=widgets.Layout(width='80%', height='120px')
)

generate_button = widgets.Button(
    description='Generate Test Cases',
    button_style='success',
    layout=widgets.Layout(width='30%')
)

output_area = widgets.Output()


# -----------------------
# BUTTON CLICK HANDLER
# -----------------------
def on_button_click(b):
    with output_area:
        clear_output(wait=True)
        print("Generating test cases using Llama 3.1...\n")
        
        user_input = input_box.value.strip()
        
        if not user_input:
            print("⚠️ Please enter a user story.")
            return
        
        try:
            results = generate_test_cases(user_input)
            clear_output(wait=True)
            print(results)
        except Exception as e:
            clear_output(wait=True)
            print("❌ ERROR:", e)


generate_button.on_click(on_button_click)


# -----------------------
# DISPLAY GUI
# -----------------------
display(input_box)
display(generate_button)
display(output_area)
```
